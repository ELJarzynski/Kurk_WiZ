# Tworzenie i modyfikacja grafu w NetworkX

```python
import networkx as nx

G = nx.Graph()

G.add_node(1)

G.add_nodes_from([2,3])

G.add_node(1)
G.add_node(2)
G.add_node(3)
G.add_node(4)
G.add_node(7)
G.add_node(9)

G.add_edge(1,2)
G.add_edge(3,1)
G.add_edge(2,4)
G.add_edge(4,1)
G.add_edge(9,1)
G.add_edge(1,7)
G.add_edge(2,9)

# To get all the nodes of a graph
node_list = G.nodes()
print("#1")
print(node_list)
```

# Operacje na krawędziach i węzłach grafu

```python
# To get all the edges of a graph
edge_list = G.edges()
print("#2")
print(edge_list)
```

```python
# To remove a node of a graph
G.remove_node(3)
node_list = G.nodes()
print("#3")
print(node_list)
```

```python
# To remove an edge of a graph
G.remove_edge(1,2)
edge_list = G.edges()
print("#4")
print(edge_list)
```

```python
# To find number of nodes
n = G.number_of_nodes()
print("#5")
print(n)
```

```python
# To find number of edges
m = G.number_of_edges()
print("#6")
print(m)
```

```python
# To find degree of a node
# d will store degree of node 2
d = G.degree(2)
print("#7")
print(d)
```

```python
# To find all the neighbor of a node
neighbor_list = G.neighbors(2)
print("#8")
print(neighbor_list)
```

# Zaawansowane operacje na grafach i wagi krawędzi

```python
edges = [(1, 2), (1, 6), (2, 3), (2, 4), (2, 6),
         (3, 4), (3, 5), (4, 8), (4, 9), (6, 7)]

G.add_edges_from(edges)
nx.draw_networkx(G, label = True)

print("Total number of nodes: ", int(G.number_of_nodes()))
print("Total number of edges: ", int(G.number_of_edges()))
print("List of all nodes: ", list(G.nodes()))
print("List of all edges: ", list(G.edges(data = True)))
print("Degree for all nodes: ", dict(G.degree()))

print("Total number of self-loops: ", int(nx.number_of_selfloops(G)))
print("List of all nodes with self-loops: ", 
      list(nx.nodes_with_selfloops(G)))

print("List of all nodes we can go to in a single step from node 2: ",
      list(G.neighbors(2)))
```

```python
# Graph with weighted edges
G = nx.Graph()

edges = [(1, 2, 19), (1, 6, 15), (2, 3, 6), (2, 4, 10),
         (2, 6, 22), (3, 4, 51), (3, 5, 14), (4, 8, 20),
         (4, 9, 42), (6, 7, 30)]

G.add_weighted_edges_from(edges)
nx.draw_networkx(G, with_labels = True)
```

# Zaawansowana wizualizacja grafu z wagami i etykietami

```python
import matplotlib.pyplot as plt
import numpy as np

G = nx.Graph()

G.add_edge("Argentina", "Chile", weight=0.6)
G.add_edge("Argentina", "Brazil", weight=0.1)
G.add_edge("Brazil", "Paraguay", weight=0.2)
G.add_edge("Brazil", "Columbia", weight=0.2)
G.add_edge("Argentina", "Paraguay", weight=0.3)
G.add_edge("Argentina", "Uruguay", weight=0.2)
G.add_edge("Argentina", "Bolivia", weight=0.2)
G.add_edge("Brazil", "Peru", weight=0.5)
G.add_edge("Brazil", "Uruguay", weight=0.1)
G.add_edge("Brazil", "Venezuela", weight=0.2)
G.add_edge("Brazil", "Bolivia", weight=0.5)
G.add_edge("Chile", "Bolivia", weight=0.2)
G.add_edge("Chile", "Peru", weight=0.1)
G.add_edge("Bolivia", "Peru", weight=0.3)
G.add_edge("Bolivia", "Paraguay", weight=0.2)
G.add_edge("Peru", "Columbia", weight=0.3)
G.add_edge("Peru", "Equador", weight=0.3)
G.add_edge("Columbia", "Equador", weight=0.3)
G.add_edge("Columbia", "Venezuela", weight=0.6)

elarge = [(u, v) for (u, v, d) in G.edges(data=True) if d["weight"] > 0.5]
esmall = [(u, v) for (u, v, d) in G.edges(data=True) if d["weight"] <= 0.5]

pos = nx.spring_layout(G, seed=7)  # positions for all nodes - seed for reproducibility

# nodes
nx.draw_networkx_nodes(G, pos, node_size=700)

# edges
nx.draw_networkx_edges(G, pos, edgelist=elarge, width=6)
nx.draw_networkx_edges(
    G, pos, edgelist=esmall, width=6, alpha=0.5, edge_color="b", style="dashed"
)

# node labels
nx.draw_networkx_labels(G, pos, font_size=20, font_family="sans-serif")
# edge weight labels
edge_labels = nx.get_edge_attributes(G, "weight")
nx.draw_networkx_edge_labels(G, pos, edge_labels)

ax = plt.gca()
ax.margins(0.08)
plt.axis("off")
plt.tight_layout()
plt.show()
```

# Najważniejsze atrybuty węzłów

```python
import networkx as nx

G = nx.Graph()

G.add_edge("Argentina", "Chile", weight=0.6)
G.add_edge("Argentina", "Brazil", weight=0.1)
G.add_edge("Brazil", "Paraguay", weight=0.2)
G.add_edge("Brazil", "Columbia", weight=0.2)
G.add_edge("Argentina", "Paraguay", weight=0.3)
G.add_edge("Argentina", "Uruguay", weight=0.2)
G.add_edge("Argentina", "Bolivia", weight=0.2)
G.add_edge("Brazil", "Peru", weight=0.5)
G.add_edge("Brazil", "Uruguay", weight=0.1)
G.add_edge("Brazil", "Venezuela", weight=0.2)
G.add_edge("Brazil", "Bolivia", weight=0.5)
G.add_edge("Chile", "Bolivia", weight=0.2)
G.add_edge("Chile", "Peru", weight=0.1)
G.add_edge("Bolivia", "Peru", weight=0.3)
G.add_edge("Bolivia", "Paraguay", weight=0.2)
G.add_edge("Peru", "Columbia", weight=0.3)
G.add_edge("Peru", "Equador", weight=0.3)
G.add_edge("Columbia", "Equador", weight=0.3)
G.add_edge("Columbia", "Venezuela", weight=0.6)

print("Node Closeness Centrality:", nx.closeness_centrality(G))
print("Node Degree Centrality:", nx.degree_centrality(G))
print("Clustering Coefficient:", nx.clustering(G))
print("Node Betweenness Centrality:", nx.betweenness_centrality(G))
print("Edge Betweenness Centrality:", nx.edge_betweenness_centrality(G))
print("Communicability:", nx.communicability(G))
print("Average Neighbor Degree:", nx.average_neighbor_degree(G))
print("Eigenvector Centrality:", nx.eigenvector_centrality(G))
```

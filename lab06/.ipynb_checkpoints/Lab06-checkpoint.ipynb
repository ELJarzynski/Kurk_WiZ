# Zadanie zmiana wykresu na badziej rozwinięty podstawowy kod jest brany z dokumentacji https://matplotlib.org/3.1.3/gallery/
## Wykres Stacked Bar Graph

```python
import numpy as np
import matplotlib.pyplot as plt

N = 5
menMeans = (20, 35, 30, 35, 27)
womenMeans = (25, 32, 34, 20, 25)
menStd = (2, 3, 4, 1, 2)
womenStd = (3, 5, 2, 3, 3)
ind = np.arange(N)    # the x locations for the groups
width = 0.35       # the width of the bars: can also be len(x) sequence

p1 = plt.bar(ind, menMeans, width, yerr=menStd)
p2 = plt.bar(ind, womenMeans, width,
             bottom=menMeans, yerr=womenStd)

plt.ylabel('Scores')
plt.title('Scores by group and gender')
plt.xticks(ind, ('G1', 'G2', 'G3', 'G4', 'G5'))
plt.yticks(np.arange(0, 81, 10))
plt.legend((p1[0], p2[0]), ('Men', 'Women'))

plt.show()
```

## Zmieniony wykres Stacked Bar Graph

```python
import numpy as np
import matplotlib.pyplot as plt

N = 5
menMeans = (20, 35, 30, 35, 27)
womenMeans = (25, 32, 34, 20, 25)
menStd = (2, 3, 4, 1, 2)
womenStd = (3, 5, 2, 3, 3)
ind = np.arange(N)    # lokalizacje etykiet
width = 0.35          # szerokość słupków

# ZMIANA 1: większa figura
fig, ax = plt.subplots(figsize=(10,6))


# ZMIANA 2: kolory i wzory słupków
p1 = ax.bar(ind, menMeans, width, yerr=menStd, 
            color='skyblue', edgecolor='black', hatch='//', label='Men')

p2 = ax.bar(ind, womenMeans, width, bottom=menMeans, yerr=womenStd, 
            color='salmon', edgecolor='black', hatch='\\\\', label='Women')

# ZMIANA 3: etykiety osi, tytuł, czcionki i kolory
ax.set_ylabel('Scores', fontsize=12, fontweight='bold', color='darkblue')
ax.set_xlabel('Groups', fontsize=12, fontweight='bold', color='darkblue')
ax.set_title('Scores by Group and Gender', fontsize=14, fontweight='bold', color='darkgreen')

# ZMIANA 4: etykiety osi X i Y – większa czcionka i kolory
ax.set_xticks(ind)
ax.set_xticklabels(('G1', 'G2', 'G3', 'G4', 'G5'), fontsize=11, color='purple')
ax.set_yticks(np.arange(0, 81, 10))
ax.tick_params(axis='y', labelsize=11, colors='darkred')

# ZMIANA 5: legenda z większą czcionką
ax.legend(fontsize=11)



# Dopasowanie layoutu i wyświetlenie
fig.tight_layout()
plt.show()
```

# Wykres 2 Plotting categorical variables

## Podstawowy wykres z dokumentacji

```python
import matplotlib.pyplot as plt

data = {'apples': 10, 'oranges': 15, 'lemons': 5, 'limes': 20}
names = list(data.keys())
values = list(data.values())

fig, axs = plt.subplots(1, 3, figsize=(9, 3), sharey=True)
axs[0].bar(names, values)
axs[1].scatter(names, values)
axs[2].plot(names, values)
fig.suptitle('Categorical Plotting')
```

# Zminiony wykres Plotting categorical variables

```python
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
data = {'apples': 10, 'oranges': 15, 'lemons': 5, 'limes': 20}
names = list(data.keys())
values = list(data.values())

# ZMIANA 1: zwiększenie rozmiaru figury
fig, axs = plt.subplots(1, 3, figsize=(12, 4), sharey=True)

# ZMIANA 2: kolory, kształty i style dla każdego wykresu
# Wykres słupkowy
axs[0].bar(names, values, color='skyblue', edgecolor='black')
axs[0].set_title('Bar Plot', fontsize=12, fontweight='bold', color='darkblue')
axs[0].set_ylabel('Values', fontsize=11, fontweight='bold', color='darkgreen')

# Wykres punktowy
axs[1].scatter(names, values, color='salmon', s=100, marker='D')  # s=100 → wielkość punktów, marker='D' → romby
axs[1].set_title('Scatter Plot', fontsize=12, fontweight='bold', color='darkblue')

# Wykres liniowy
axs[2].plot(names, values, color='purple', linewidth=2, marker='o', markersize=8)  # pogrubiona linia + marker
axs[2].set_title('Line Plot', fontsize=12, fontweight='bold', color='darkblue')

# ZMIANA 3: tytuł główny figury
fig.suptitle('Categorical Plotting', fontsize=14, fontweight='bold', color='darkred')

# ZMIANA 4: etykiety osi X w większej czcionce
for ax in axs:
    ax.set_xticklabels(names, fontsize=11, color='purple')

fig.tight_layout(rect=[0, 0, 1, 0.95])  # Dopasowanie layoutu z miejscem na suptitle
plt.show()
```

# Wykres 3 Horizontal bar chart

## Podstawowy wykres z dokumentacji

```python
import matplotlib.pyplot as plt
import numpy as np

# Fixing random state for reproducibility
np.random.seed(19680801)


plt.rcdefaults()
fig, ax = plt.subplots()

# Example data
people = ('Tom', 'Dick', 'Harry', 'Slim', 'Jim')
y_pos = np.arange(len(people))
performance = 3 + 10 * np.random.rand(len(people))
error = np.random.rand(len(people))

ax.barh(y_pos, performance, xerr=error, align='center')
ax.set_yticks(y_pos)
ax.set_yticklabels(people)
ax.invert_yaxis()  # labels read top-to-bottom
ax.set_xlabel('Performance')
ax.set_title('How fast do you want to go today?')

plt.show()
```

# Zminiony wykres Horizontal bar chart

```python
import matplotlib.pyplot as plt
import numpy as np

# Fixing random state for reproducibility
np.random.seed(19680801)

plt.rcdefaults()

# ZMIANA 1: zwiększenie rozmiaru figury
fig, ax = plt.subplots(figsize=(10,6))

# Example data
people = ('Tom', 'Dick', 'Harry', 'Slim', 'Jim')
y_pos = np.arange(len(people))
performance = 3 + 10 * np.random.rand(len(people))
error = np.random.rand(len(people))

# ZMIANA 2: kolory słupków i wzory
bars = ax.barh(y_pos, performance, xerr=error, align='center',
               color='skyblue', edgecolor='black', hatch='//')

# ZMIANA 3: etykiety osi, tytuł, czcionki i kolory
ax.set_xlabel('Performance', fontsize=12, fontweight='bold', color='darkblue')
ax.set_title('How fast do you want to go today?', fontsize=14, fontweight='bold', color='darkgreen')

# ZMIANA 4: etykiety Y z większą czcionką i kolorem
ax.set_yticks(y_pos)
ax.set_yticklabels(people, fontsize=11, color='purple')
ax.invert_yaxis()  # labels read top-to-bottom

# ZMIANA 5: dodanie wartości na końcu słupków
for bar in bars:
    width = bar.get_width()
    ax.annotate(f'{width:.2f}',
                xy=(width, bar.get_y() + bar.get_height() / 2),
                xytext=(3, 0),  # offset
                textcoords="offset points",
                ha='left', va='center',
                fontsize=10, fontweight='bold', color='black')

# Dopasowanie layoutu i wyświetlenie
fig.tight_layout()
plt.show()
```

# Wykres 4 Grouped bar chart with labels

## Podstawowy wykres z dokumentacji

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np


labels = ['G1', 'G2', 'G3', 'G4', 'G5']
men_means = [20, 34, 30, 35, 27]
women_means = [25, 32, 34, 20, 25]

x = np.arange(len(labels))  # the label locations
width = 0.35  # the width of the bars

fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, men_means, width, label='Men')
rects2 = ax.bar(x + width/2, women_means, width, label='Women')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Scores')
ax.set_title('Scores by group and gender')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()


def autolabel(rects):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')


autolabel(rects1)
autolabel(rects2)

fig.tight_layout()

plt.show()
```

# Zminiony wykres Horizontal bar chart

```python
import matplotlib.pyplot as plt
import numpy as np


labels = ['G1', 'G2', 'G3', 'G4', 'G5']
men_means = [20, 34, 30, 35, 27]
women_means = [25, 32, 34, 20, 25]

x = np.arange(len(labels))  # the label locations
width = 0.35  # the width of the bars

# ZMIANA 1: zwiększenie rozmiaru całej figury
fig, ax = plt.subplots(figsize=(10,6))  # większa figura

# ZMIANA 2: kolory słupków, kształty (wzory)
rects1 = ax.bar(x - width/2, men_means, width, 
                label='Men', 
                color='skyblue',           # zmiana koloru słupka
                edgecolor='black',         # obramowanie słupka
                hatch='//')                # wzór w słupku

rects2 = ax.bar(x + width/2, women_means, width, 
                label='Women', 
                color='salmon',            # zmiana koloru słupka
                edgecolor='black',         # obramowanie
                hatch='\\\\')              # inny wzór w słupku

# ZMIANA 3: etykiety i tytuł z większą czcionką, pogrubione, kolorowe
ax.set_ylabel('Scores', fontsize=12, fontweight='bold', color='darkblue')  # oś Y
ax.set_xlabel('Groups', fontsize=12, fontweight='bold', color='darkblue')  # oś X
ax.set_title('Scores by Group and Gender', fontsize=14, fontweight='bold', color='darkgreen')  # tytuł


# ZMIANA 4: etykiety osi X – większa czcionka, kolor
ax.set_xticks(x)
ax.set_xticklabels(labels, fontsize=11, rotation=0, color='purple')


# legenda z większą czcionką
ax.legend(fontsize=11)


# ZMIANA 5: wartości nad słupkami – pogrubione
def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # offset w pionie
                    textcoords="offset points",
                    ha='center', va='bottom',
                    fontsize=10, fontweight='bold', color='black')  # pogrubiony tekst

autolabel(rects1)
autolabel(rects2)

fig.tight_layout()
plt.show()
```

```python
import pandas as pd

df = pd.read_excel("Wine_data_prep.xlsx")
print(df.head())
```

```python
# Sprawdź, czy dane są uzupełnione w wierszach i kolumnach.
print("Braki przed uzupełnieniem:")
print(df.isnull().sum())
```

```python
# Puste komórki wypełniają się średnimi wartościami z komórek sąsiada.
df.fillna(df.select_dtypes(include='number').mean(), inplace=True)
print("Braki po uzupełnieniu intów:")
print(df.isnull().sum())
```

```python
# Czyli wine jest taktowane jako kolumna string

# Czyszczenie nazw kolumn ze spacji
df.columns = df.columns.str.strip()

# Konwersja stringów w Nany
for col in df.columns:
    df[col] = pd.to_numeric(df[col], errors='coerce') 
    
print(df.tail())
```

```python
df.fillna(df.select_dtypes(include='number').mean(), inplace=True)
print("Braki po uzupełnienieniu:")
print(df.isnull().sum())
print(df.tail())
```

```python
df.to_csv("wine_data_c.csv", index=False, encoding="cp1250", sep=";")
```
